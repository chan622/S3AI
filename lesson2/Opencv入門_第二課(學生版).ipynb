```python
# 例子一：利用 np.zeros() 建立圖像矩陣
import numpy as np
import cv2

# 建立一張 256x256 的 RGB 圖片（黑色）
# 形式：np.zeros(shape, dtype=float, order=’C’)
# shape: 形狀，(高, 寬, 每單位內含個數)
# dtype: 資料型別，可選引數，預設numpy.float64
# order: 可選引數，(默認)「C」代表與C語言類似，行優先；「F」代表列優先
# uint8: 8 位无符号整数（0 ~ 255）

img = np.zeros((256, 256, 3), np.uint8)
print(img)

cv2.imshow('sample_2_1',img)
# cv2.imwrite('sample_2_1.jpg',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 利用 fill() 進行填色

# 電腦上三原色的顯示範圍為0到255
# RGB為（255, 0, 0）表示全紅, (0, 255, 0)表示全綠, (0, 0, 255)表示全藍
# (0, 0, 0)表示黑色, （255, 255, 255)表示白色
# 將圖片用淺灰色 (200, 200, 200) 填滿
# 形式：img.fill(0~255)

img.fill(200)

cv2.imshow('sample_2_1a',img)
# cv2.imwrite('sample_2_1a.jpg',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Mac機 / 如上一塊代碼執行不了，可執行這塊代碼
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
img = np.zeros((256, 256, 3), np.uint8)

img.fill(200)

plt.imshow(img)
```

```python
# 利用 opencv 加入圖形

# shape 座標圖：
#       0            255  —>  先記 橫坐標
#  0   [             ]
#      [             ]
#      [             ]
#      [             ]
#      [             ]
# 255  [             ]
#  |
#  V
# 後記 縱坐標

# 在圖片上畫一條紅色的對角線，寬度為 5 px
# 形式: cv2.line(影像, 開始座標, 結束座標, 顏色, 線條寬度)
cv2.line(img, (0, 0), (255, 255), (255, 255, 187), 10)
cv2.imshow('sample_2_1b',img)
# cv2.imwrite('sample_2_1b.jpg',img)

# 在圖片上畫一條綠色的箭頭線段，寬度為 5 px
# 形式: cv2.line(影像, 開始座標, 結束座標, 顏色, 線條寬度)
img = cv2.arrowedLine(img, (255, 0), (0, 255), (0, 255, 0), 5)
cv2.imshow('sample_2_1c',img)
# cv2.imwrite('sample_2_1c.jpg',img)


# 在圖片上畫一條藍色的矩形，寬度為 1 px
# 形式: cv2.rectangle(影像, 開始座標, 結束座標, 顏色, 線條寬度，若設定為-1，則代表實心)
img = cv2.rectangle(img, (85,85), (170,170), (255,0,0), 1)
cv2.imshow('sample_2_1d',img)
# cv2.imwrite('sample_2_1d.jpg',img)


# 在圖片上畫一條黑色的圓形，寬度為 1 px
# 形式: cv2.circle(影像, 圓心座標, 半徑, 顏色, 線條寬度，若設定為-1，則代表實心)
img = cv2.circle(img, (128,128), 100, (0,0,0), 1)
cv2.imshow('sample_2_1e',img)
# cv2.imwrite('sample_2_1e.jpg',img)

# 畫橢圓形
# cv2.ellipse(影像, 中心座標, 軸長, 旋轉角度, 起始角度, 結束角度, 顏色, 線條寬度)

# 按下任意鍵則關閉所有視窗
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 例子二：利用 cv2.putText() 加入文字

import numpy as np
import cv2

img = np.zeros((400, 400, 3), np.uint8)
img.fill(90)

# 自定文字
text = 'Hello, OpenCV!'
# text = '你好, OpenCV!'

# 使用各種字體
# cv2.putText(影像, 文字, 座標, 字型, 大小, 顏色, 線條寬度, 線條種類)
cv2.putText(img, text, (10, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 1, cv2.LINE_AA)

cv2.putText(img, text, (10, 80), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1, cv2.LINE_AA)

cv2.putText(img, text, (10, 120), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 255, 255), 1, cv2.LINE_AA)

cv2.putText(img, text, (10, 160), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 255), 1, cv2.LINE_AA)

cv2.putText(img, text, (10, 200), cv2.FONT_HERSHEY_TRIPLEX, 1, (0, 255, 255), 1, cv2.LINE_AA)

cv2.putText(img, text, (10, 240), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 255, 255), 1, cv2.LINE_AA)

cv2.putText(img, text, (10, 280), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 1, (0, 255, 255), 1, cv2.LINE_AA)

cv2.putText(img, text, (10, 320), cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 1, (0, 255, 255), 1, cv2.LINE_AA)

cv2.imshow('sample_2_2', img)
# cv2.imwrite('sample_2_2.jpg',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 例子三：利用 cv2.setMouseCallback() 自定義按鍵事件

import cv2
import numpy as np

def click_event(event, x, y, flags, param):
    # 按下左鍵時，則加入(x,y)座標文字
    if event == cv2.EVENT_LBUTTONDOWN:
        print('座標:(',x,',',y,')',sep='')
        font = cv2.FONT_HERSHEY_SIMPLEX
        strXY = str(x) + ', ' + str(y)
        cv2.putText(img, strXY, (x,y), font, 0.3, (0,0,0), 1, cv2.LINE_AA)
        cv2.imshow('sample_2_3', img)
    # 按下右鍵時，則加入RGB色碼文字
    if event == cv2.EVENT_RBUTTONDOWN:
        blue = img[y,x,0]
        green = img[y,x,1]
        red = img[y,x,2]
        font = cv2.FONT_HERSHEY_SIMPLEX
        strBGR = str(blue) + ', ' + str(green) + ', ' + str(red)
        print('GRB色碼:(',strBGR,')',sep='')
        cv2.putText(img, strBGR, (x,y), font, 0.3, (0,0,255), 1, cv2.LINE_AA)
        cv2.imshow('sample_2_3', img)

img = cv2.imread('pic3.png', 1)
cv2.imshow('sample_2_3', img)

cv2.setMouseCallback('sample_2_3', click_event)

a = cv2.waitKey(0) & 0xFF
if a == ord('q'):
    cv2.destroyAllWindows()
elif a == ord('s'):
    cv2.imwrite('sample_2_3.jpg', img)
    cv2.destroyAllWindows()
```
